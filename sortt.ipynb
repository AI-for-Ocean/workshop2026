```python
import json
```

```python
with open("src/data/bios.json") as f:
    data = json.load(f)

with open("c:/users/krish/Downloads/bios_final.json") as f:
    roles_data = json.load(f)
```

```python
for person in data:
    for role_person in roles_data:
        if person["name"] == role_person["name"]:
            person["role"] = role_person["role"]
```

```python
with open("src/data/bios.json", "w") as f:
    json.dump(data, f, indent=2)
```
